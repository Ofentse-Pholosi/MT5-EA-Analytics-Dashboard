```python
import pandas as pd

file_path = "../data/raw/ReportHistory-40892819.xlsx"

# Load entire sheet without assuming headers
raw_df = pd.read_excel(file_path, header=None)

raw_df.head(20)
```

```python
raw_df[raw_df.iloc[:, 0] == "Deals"]
```

```python
deals_index = raw_df[raw_df.iloc[:, 0] == "Deals"].index[0]

raw_df.iloc[deals_index - 5 : deals_index + 10]
```

```python
# Identify the row index where the Deals table begins
deals_start_row = raw_df[raw_df.iloc[:, 0] == "Deals"].index[0]

deals_start_row
```

```python
# Extract column headers for the Deals table
headers = raw_df.iloc[deals_start_row + 1].tolist()

headers
```

```python
# Extract all rows below the header row as deal records
deals_df = raw_df.iloc[deals_start_row + 2:].copy()

# Assign the extracted headers as column names
deals_df.columns = headers
```

```python
# Remove rows that are completely empty
deals_df = deals_df.dropna(how="all")
```

```python
# Preview the cleaned Deals dataset
deals_df.head()
```

```python
# Convert numeric columns to numeric data types
numeric_cols = [
    "Volume", "Price", "Commission",
    "Fee", "Swap", "Profit", "Balance"
]

for col in numeric_cols:
    deals_df[col] = pd.to_numeric(deals_df[col], errors="coerce")

# Convert time column to datetime
deals_df["Time"] = pd.to_datetime(deals_df["Time"])
```

```python
# Identify rows where Time is not a valid timestamp
deals_df[~deals_df["Time"].astype(str).str.match(r"\d{4}\.\d{2}\.\d{2}")]
```

```python
# Keep only rows where Time looks like a proper MT5 timestamp
deals_df = deals_df[
    deals_df["Time"].astype(str).str.match(r"\d{4}\.\d{2}\.\d{2}")
]
```

```python
# Convert Time column to datetime after cleaning
deals_df["Time"] = pd.to_datetime(
    deals_df["Time"],
    format="%Y.%m.%d %H:%M:%S"
)
```

```python
# Separate balance operations
deposits_df = deals_df[deals_df["Type"] == "balance"]

# Separate trades
trades_df = deals_df[deals_df["Type"] == "trade"]

# Separate entries and exits
entries_df = trades_df[trades_df["Direction"] == "in"]
exits_df   = trades_df[trades_df["Direction"] == "out"]
```

```python
print("Total realized profit:", exits_df["Profit"].sum())
print("Final account balance:", deals_df["Balance"].dropna().iloc[-1])
```

```python
exits_df = deals_df[
    (deals_df["Type"] == "trade") &
    (deals_df["Profit"] != 0)
]
```

```python
# Identify exit trades based on observed MT5 encoding
exits_df = deals_df[
    (deals_df["Direction"] == "out") &
    (deals_df["Type"].isin(["buy", "sell"]))
]
```

```python
print("Exit trades count:", len(exits_df))
print("Total realized profit:", exits_df["Profit"].sum())
```

```python
# Separate deposits / balance operations
deposits_df = deals_df[deals_df["Type"] == "balance"]
```

```python
# Total number of exit trades
total_trades = len(exits_df)

# Number of winning and losing trades
winning_trades = exits_df[exits_df["Profit"] > 0]
losing_trades  = exits_df[exits_df["Profit"] < 0]

print("Total trades:", total_trades)
print("Winning trades:", len(winning_trades))
print("Losing trades:", len(losing_trades))
```

```python
# Win rate as a percentage
win_rate = len(winning_trades) / total_trades * 100

print(f"Win rate: {win_rate:.2f}%")
```

```python
# Average profit per winning trade
avg_win = winning_trades["Profit"].mean()

# Average loss per losing trade (absolute value)
avg_loss = abs(losing_trades["Profit"].mean())

print("Average win:", round(avg_win, 2))
print("Average loss:", round(avg_loss, 2))
```

```python
# Expectancy calculation
expectancy = (
    (win_rate / 100) * avg_win
    - (1 - win_rate / 100) * avg_loss
)

print("Expectancy per trade:", round(expectancy, 2))
```

```python
# Quick statistical overview of trade profits
exits_df["Profit"].describe()
```

```python
# Create a date column for grouping
exits_df["Date"] = exits_df["Time"].dt.date

# Trades per day
trades_per_day = exits_df.groupby("Date").size()

trades_per_day
```

```python
# Create a safe copy before modifying
exits_df = exits_df.copy()

# Add derived Date column explicitly
exits_df.loc[:, "Date"] = exits_df["Time"].dt.date
```

```python
# Calculate daily profit per instrument
daily_profit_symbol = (
    exits_df
    .groupby(["Date", "Symbol"])["Profit"]
    .sum()
    .reset_index()
)

daily_profit_symbol.head()
```

```python
DAILY_TARGET = 50

# Flag whether daily target was reached per instrument
daily_profit_symbol["Target_Hit"] = (
    daily_profit_symbol["Profit"] >= DAILY_TARGET
)
```

```python
# Trades per instrument per day
trades_per_symbol_day = (
    exits_df
    .groupby(["Date", "Symbol"])
    .size()
    .reset_index(name="Trade_Count")
)

trades_per_symbol_day.head()
```

```python
# Sort for proper cumulative calculation
exits_df = exits_df.sort_values("Time")

# Cumulative profit per instrument
exits_df["Equity"] = (
    exits_df
    .groupby("Symbol")["Profit"]
    .cumsum()
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

for symbol in exits_df["Symbol"].unique():
    symbol_df = exits_df[exits_df["Symbol"] == symbol]
    plt.plot(symbol_df["Time"], symbol_df["Equity"], label=symbol)

plt.title("Equity Curves per Instrument")
plt.xlabel("Time")
plt.ylabel("Cumulative Profit")
plt.legend()
plt.grid(True)
plt.show()
```
